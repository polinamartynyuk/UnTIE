```python
def get_qa_attention_weights(question, context):
    """
    Визуализирует веса внимания модели RoBERTa для QA между вопросом и контекстом.
    
    Args:
        question (str): Текст вопроса
        context (str): Текст контекста для поиска ответа
        model_name (str): Название предобученной модели (по умолчанию 'deepset/roberta-base-squad2')
    
    Returns:
        dict: Словарь с токенами и их весами внимания
    """
    
    # Загрузка модели и токенизатора
    model = BERTEngQASingleton()

    tokenizer = model.tokenizer
    model = model.model
    
    # Токенизация
    inputs = tokenizer(question, context, return_tensors='pt', truncation=True, max_length=512)
    inputs = {key: value.to(device) for key, value in inputs.items()}
    input_ids = inputs['input_ids']
    
    # Получение предсказаний модели с вниманием
    with torch.no_grad():
        outputs = model(**inputs)
        attentions = outputs.attentions  # Все слои внимания
    
    # Среднее внимание по всем головам и слоям
    all_attentions = torch.stack(attentions)  # [layers, batch, heads, seq_len, seq_len]
    averaged_attention = all_attentions.mean(dim=(0, 1, 2))  # [seq_len, seq_len]
    
    # Индексы токенов контекста (после [SEP])
    sep_index = torch.where(input_ids[0] == tokenizer.sep_token_id)[0][0].item()
    context_start = sep_index + 1
    context_end = len(input_ids[0]) - 1  # Исключаем конечный [SEP]
    
    # Извлекаем внимание к токенам контекста
    context_attention = averaged_attention[context_start:context_end+1, context_start:context_end+1]
    
    # Агрегируем внимание по строкам (входящие связи)
    token_importance = context_attention.mean(dim=0)
    
    # Нормализация
    token_importance = token_importance / token_importance.sum()
    
    # Сопоставление с токенами
    context_tokens = tokenizer.convert_ids_to_tokens(input_ids[0][context_start:context_end+1])
    
    return {
        'tokens': context_tokens,
        'attention_weights': token_importance.cpu().numpy(),
        'question': question,
        'context': context
    }
```

```python
def clean_token(token):
    """Очищает токен от специальных символов RoBERTa"""
    # Убираем символ пробела Ġ
    cleaned = token.replace('Ġ', ' ')
    # Убираем лишние пробелы
    cleaned = cleaned.strip()
    # Убираем точки, которые являются частью токенизации (но не в середине слова)
    if cleaned.endswith('.') and len(cleaned) > 1:
        cleaned = cleaned[:-1]
    # Приводим к нижнему регистру
    cleaned = cleaned.lower()
    return cleaned

def reconstruct_words(tokens, weights):
    """Восстанавливает полные слова из субтокенов"""
    words = []
    word_weights = []
    current_word = ""
    current_weight = 0
    count = 0
    
    for token, weight in zip(tokens, weights):
        cleaned_token = clean_token(token)
        
        # Пропускаем пустые токены после очистки
        if not cleaned_token:
            continue
            
        # Если токен начинается с пробела - это новое слово
        if token.startswith('Ġ') and current_word:
            # Убираем возможную точку в конце слова
            if current_word.endswith('.'):
                current_word = current_word[:-1]
            words.append(current_word)
            word_weights.append(current_weight / count if count > 0 else 0)
            current_word = cleaned_token
            current_weight = weight
            count = 1
        else:
            # Продолжение текущего слова (субтокен)
            current_word += cleaned_token
            current_weight += weight
            count += 1
    
    # Добавляем последнее слово
    if current_word:
        # Убираем возможную точку в конце последнего слова
        if current_word.endswith('.'):
            current_word = current_word[:-1]
        words.append(current_word)
        word_weights.append(current_weight / count if count > 0 else 0)
    
    return words, np.array(word_weights)

def filter_special_tokens(tokens, weights):
    """Фильтрует специальные токены"""
    valid_tokens = []
    valid_weights = []
    
    special_tokens = ['<s>', '</s>', '<pad>', '[CLS]', '[SEP]', '.', '?', '!', ',', '/', '//', '][']
    
    for token, weight in zip(tokens, weights):
        # Пропускаем специальные токены и знаки препинания
        if (token not in special_tokens and 
            not token.startswith('<') and 
            not token.endswith('>') and
            len(token.strip('.')) > 0):  # Исключаем токены, состоящие только из точек
            valid_tokens.append(token)
            valid_weights.append(weight)
    
    return valid_tokens, np.array(valid_weights)
```

```python
def visualize_top_attention_clean(tokens, weights, top_k=10):
    """Визуализация с очищенными токенами"""
    # Фильтруем специальные токены
    filtered_tokens, filtered_weights = filter_special_tokens(tokens, weights)
    
    # Восстанавливаем полные слова
    words, word_weights = reconstruct_words(filtered_tokens, filtered_weights)
    
    # Нормализуем веса после восстановления
    # if len(word_weights) > 0:
    #    word_weights = word_weights / word_weights.sum()
    
    indices = np.argsort(word_weights)[::-1][:top_k]
    
    print("Топ-%d наиболее важных слов:" % top_k)
    for i, idx in enumerate(indices):
        if i < len(words):
            print(f"{i+1}. '{words[idx]}': {word_weights[idx]:.4f}")
```

```python
def keywords_top_attention_clean(tokens, weights, top_k=10):
    """Визуализация с очищенными токенами"""
    # Фильтруем специальные токены
    filtered_tokens, filtered_weights = filter_special_tokens(tokens, weights)
    
    # Восстанавливаем полные слова
    words, word_weights = reconstruct_words(filtered_tokens, filtered_weights)
    
    # Нормализуем веса после восстановления
    # if len(word_weights) > 0:
    #    word_weights = word_weights / word_weights.sum()
    
    indices = np.argsort(word_weights)[::-1][:top_k]

    keywords = []

    for i, idx in enumerate(indices):
        if i < len(words):
            keywords.append({
                'word': words[idx],
                'weight': word_weights[idx]
            })

    return keywords
```

```python
def postprocess_words_array(words_dicts: list) -> list:
    """
    Постобработка массива слов: удаление дубликатов (оставляем слово с наивысшим весом)
    и удаление стоп-слов английского языка.
    """
    # Загружаем стоп-слова
    stop_words = {
        'the', 'a', 'an', 'and', 'or', 'but', 'in', 'on', 'at', 'to', 'for', 'of', 'with', 'by',
        'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'do', 'does',
        'did', 'will', 'would', 'could', 'should', 'may', 'might', 'must', 'can', 'i', 'you',
        'he', 'she', 'it', 'we', 'they', 'me', 'him', 'her', 'us', 'them', 'this', 'that',
        'these', 'those', 'my', 'your', 'his', 'her', 'its', 'our', 'their', 'as', 'from'
    }
    
    # Словарь для хранения максимального веса для каждого слова
    word_max_weight = {}
    
    # Находим максимальный вес для каждого слова
    for word_dict in words_dicts:
        word = word_dict['word'].lower()
        weight = word_dict['weight']
        
        if word not in word_max_weight or weight > word_max_weight[word]:
            word_max_weight[word] = weight
    
    # Создаем новый массив без дубликатов и стоп-слов
    result = []
    for word, max_weight in word_max_weight.items():
        if word not in stop_words and len(word) > 1:  # исключаем стоп-слова и одиночные символы
            result.append({"word": word, "weight": max_weight})
    
    # Сортируем по весу в убывающем порядке
    result.sort(key=lambda x: x['weight'], reverse=True)
    
    return result
```

```python
from typing import List, Dict
import re
from dataclasses import dataclass
import numpy as np



def score_chunks_by_keywords(
    chunks: List[TextChunk], 
    keywords: List[dict],
    case_sensitive: bool = False,
    use_tfidf_weights: bool = True,
    min_matches: int = 1
) -> List[ScoredChunk]:
    """
    Оценивает фрагменты на основе количества и важности ключевых слов.
    Чем больше ключевых слов и чем они важнее, тем выше оценка фрагмента.
    
    Параметры:
        chunks: Список объектов TextChunk для оценки
        keywords: Список словарей формата [{"word": str, "lemma": str, "stem": str, "score": float}, ...]
        case_sensitive: Учитывать регистр при поиске
        use_tfidf_weights: Использовать веса ключевых слов из поля 'score'
        min_matches: Минимальное количество совпадений для включения фрагмента в результат
    
    Возвращает:
        Список объектов ScoredChunk с оценками и информацией о совпадениях
    """
    if not keywords or not chunks:
        return []
    
    # Подготавливаем ключевые слова и их веса
    keyword_weights = {}
    search_terms = set()
    term_to_keyword = {}
    
    for keyword in keywords:
        word = keyword['word']
        weight = keyword.get('score', 1.0) if use_tfidf_weights else 1.0
        
        keyword_weights[word] = weight
        
        # Добавляем лемму и стем для поиска
        lemma = keyword.get('lemma', word)
        stem = keyword.get('stem', word)
        
        search_terms.add(lemma)
        search_terms.add(stem)
        
        term_to_keyword[lemma.lower()] = word
        term_to_keyword[stem.lower()] = word
    
    # Создаем regex-паттерн для поиска
    pattern = re.compile(
        r'\b(?:{})\b'.format('|'.join(map(re.escape, search_terms))),
        flags=0 if case_sensitive else re.IGNORECASE
    )
    
    scored_chunks = []
    
    for chunk in chunks:
        # Находим все совпадения терминов в тексте
        text_to_search = chunk.text if case_sensitive else chunk.text.lower()
        found_terms = pattern.findall(text_to_search)
        
        # Собираем уникальные найденные ключевые слова и их веса
        matched_keywords = set()
        total_score = 0.0
        
        for term in found_terms:
            normalized_term = term if case_sensitive else term.lower()
            if normalized_term in term_to_keyword:
                keyword_word = term_to_keyword[normalized_term]
                if keyword_word not in matched_keywords:
                    matched_keywords.add(keyword_word)
                    total_score += keyword_weights.get(keyword_word, 1.0)
        
        # Применяем дополнительные факторы для улучшения оценки
        if matched_keywords:
            # Нормализуем по количеству слов в фрагменте (чтобы избежать bias к длинным текстам)
            normalization_factor = 1.0 + (len(matched_keywords) / max(1, chunk.word_token_count / 10))
            
            # Учитываем плотность ключевых слов
            density_bonus = min(2.0, 1.0 + (len(matched_keywords) / max(1, len(chunk.text.split()) / 20)))
            
            final_score = total_score * normalization_factor * density_bonus
            
            if len(matched_keywords) >= min_matches:
                scored_chunks.append(ScoredChunk(
                    chunk=chunk,
                    score=final_score,
                    matched_keywords=list(matched_keywords),
                    keyword_scores={kw: keyword_weights.get(kw, 1.0) for kw in matched_keywords}
                ))
    
    # Сортируем по убыванию оценки
    scored_chunks.sort(key=lambda x: x.score, reverse=True)
    
    return scored_chunks

# Альтернативная версия с более сложной системой оценки
def score_chunks_advanced(
    chunks: List[TextChunk], 
    keywords: List[dict],
    case_sensitive: bool = False,
    min_matches: int = 1,
    position_weight: float = 0.3,
    frequency_weight: float = 0.7
) -> List[ScoredChunk]:
    """
    Расширенная версия оценки фрагментов с учетом позиции и частоты ключевых слов.
    """
    if not keywords or not chunks:
        return []
    
    # Подготавливаем ключевые слова
    keyword_weights = {}
    search_terms = set()
    term_to_keyword = {}
    
    for keyword in keywords:
        word = keyword['word']
        weight = keyword.get('score', 1.0)
        
        keyword_weights[word] = weight
        
        lemma = keyword.get('lemma', word)
        stem = keyword.get('stem', word)
        
        search_terms.add(lemma)
        search_terms.add(stem)
        
        term_to_keyword[lemma.lower()] = word
        term_to_keyword[stem.lower()] = word
    
    pattern = re.compile(
        r'\b(?:{})\b'.format('|'.join(map(re.escape, search_terms))),
        flags=0 if case_sensitive else re.IGNORECASE
    )
    
    scored_chunks = []
    
    for chunk in chunks:
        text_to_search = chunk.text if case_sensitive else chunk.text.lower()
        found_matches = list(pattern.finditer(text_to_search))
        
        if not found_matches:
            continue
        
        # Анализируем совпадения
        matched_keywords = set()
        position_scores = []
        keyword_frequencies = {}
        
        for match in found_matches:
            term = match.group()
            normalized_term = term if case_sensitive else term.lower()
            
            if normalized_term in term_to_keyword:
                keyword_word = term_to_keyword[normalized_term]
                matched_keywords.add(keyword_word)
                
                # Оценка позиции (слова в начале текста более важны)
                position = match.start() / max(1, len(text_to_search))
                position_score = 1.0 - position  # чем раньше, тем выше оценка
                position_scores.append(position_score)
                
                # Подсчет частоты
                keyword_frequencies[keyword_word] = keyword_frequencies.get(keyword_word, 0) + 1
        
        if len(matched_keywords) >= min_matches:
            # Вычисляем общую оценку
            base_score = sum(keyword_weights.get(kw, 1.0) for kw in matched_keywords)
            
            # Учитываем позицию (средняя оценка позиций)
            avg_position_score = np.mean(position_scores) if position_scores else 0.5
            position_contribution = 1.0 + position_weight * avg_position_score
            
            # Учитываем частоту (логарифмическая шкала для избежания перекоса)
            freq_contribution = 1.0 + frequency_weight * np.log1p(sum(keyword_frequencies.values()))
            
            # Учитываем уникальность (количество разных ключевых слов)
            uniqueness_bonus = 1.0 + (len(matched_keywords) / len(keywords)) * 0.5
            
            final_score = base_score * position_contribution * freq_contribution * uniqueness_bonus
            
            scored_chunks.append(ScoredChunk(
                chunk=chunk,
                score=final_score,
                matched_keywords=list(matched_keywords),
                keyword_scores=keyword_frequencies
            ))
    
    scored_chunks.sort(key=lambda x: x.score, reverse=True)
    return scored_chunks

# Функция для комбинированной фильтрации и оценки
def filter_and_score_chunks(
    chunks: List[TextChunk], 
    keywords: List[dict],
    min_keywords: int = 1,
    case_sensitive: bool = False,
    score_threshold: float = 0.0
) -> List[ScoredChunk]:
    """
    Комбинированная функция: сначала фильтрует, затем оценивает фрагменты.
    """
    # Сначала фильтруем по минимальному количеству ключевых слов
    filtered_chunks = filter_chunks_by_keywords_lemm_stemm(
        chunks, keywords, min_keywords, case_sensitive
    )
    
    # Затем оцениваем отфильтрованные фрагменты
    scored_chunks = score_chunks_by_keywords(
        filtered_chunks, keywords, case_sensitive
    )
    
    # Применяем порог оценки
    if score_threshold > 0:
        scored_chunks = [sc for sc in scored_chunks if sc.score >= score_threshold]
    
    return scored_chunks
```

```python
from typing import List, Dict, Tuple, Union
from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.cluster import DBSCAN, OPTICS
from sklearn.metrics import pairwise_distances
from collections import defaultdict

class AnswerGrouper:
    def __init__(self):
        self.model = SentenceTokenizerSingleton()
    
    def get_embeddings(self, answers):
        """Получает эмбеддинги для всех ответов"""
        texts = [answer.text for answer in answers]
        return self.model.tokenizer.encode(texts)
    
    def adaptive_clustering(self, embeddings, answers, min_samples=2, eps=0.5):
        """
        Адаптивная кластеризация ответов с использованием DBSCAN
        Возвращает словарь кластеров
        """
        # Вычисляем попарные расстояния
        distances = pairwise_distances(embeddings, metric='cosine')
        
        # Используем DBSCAN для кластеризации
        clustering = DBSCAN(eps=eps, min_samples=min_samples, metric='precomputed').fit(distances)
        labels = clustering.labels_
        
        # Группируем ответы по кластерам
        clusters = defaultdict(list)
        for i, label in enumerate(labels):
            if label != -1:  # Игнорируем выбросы
                clusters[label].append((answers[i], embeddings[i]))
        
        return dict(clusters)
    
    def hierarchical_grouping(self, embeddings, answers, threshold_percentile=75):
        """
        Иерархическая группировка с динамическим определением порога
        Возвращает список групп
        """
        # Вычисляем косинусные расстояния
        distances = pairwise_distances(embeddings, metric='cosine')
        
        # Определяем порог на основе процентиля расстояний
        flat_distances = distances[np.triu_indices_from(distances, k=1)]
        threshold = np.percentile(flat_distances, threshold_percentile)
        
        # Создаем группы на основе порога
        groups = []
        used_indices = set()
        
        for i in range(len(embeddings)):
            if i in used_indices:
                continue
                
            # Находим все похожие ответы
            group_indices = np.where(distances[i] <= threshold)[0]
            group_answers = [answers[j] for j in group_indices]
            group_embeddings = embeddings[group_indices]
            
            groups.append({
                'answers': group_answers,
                'embeddings': group_embeddings,
                'size': len(group_answers)
            })
            used_indices.update(group_indices)
        
        return groups
    
    def cluster_with_quality_assessment(self, embeddings, answers, min_cluster_size=2):
        """
        Кластеризация с оценкой качества кластеров
        Возвращает словарь кластеров и словарь качеств
        """
        # Используем OPTICS для кластеризации с автоматическим определением параметров
        clustering = OPTICS(min_samples=min_cluster_size, metric='cosine').fit(embeddings)
        labels = clustering.labels_
        
        # Оцениваем качество каждого кластера
        clusters = defaultdict(list)
        cluster_qualities = {}
        
        for i, label in enumerate(labels):
            if label != -1:
                clusters[label].append((answers[i], embeddings[i]))
        
        # Вычисляем качество кластера на основе когерентности
        for label, cluster_data in clusters.items():
            answers_in_cluster, embeddings_in_cluster = zip(*cluster_data)
            
            # Вычисляем когерентность кластера (среднее попарное сходство)
            if len(embeddings_in_cluster) > 1:
                cluster_distances = pairwise_distances(embeddings_in_cluster, metric='cosine')
                coherence = 1 - np.mean(cluster_distances)
                cluster_qualities[label] = coherence
            else:
                cluster_qualities[label] = 0
        
        return dict(clusters), cluster_qualities
    
    def group_answers(self, answers, method='adaptive', **kwargs):
        """
        Основная функция для группировки ответов и вычисления средних оценок
        """
        # Получаем эмбеддинги для ответов
        embeddings = self.get_embeddings(answers)
        
        if method == 'adaptive':
            clusters_dict = self.adaptive_clustering(embeddings, answers, **kwargs)
            clusters = self._process_cluster_dict(clusters_dict)
        elif method == 'hierarchical':
            clusters = self.hierarchical_grouping(embeddings, answers, **kwargs)
        elif method == 'quality':
            clusters_dict, qualities = self.cluster_with_quality_assessment(embeddings, answers, **kwargs)
            clusters = self._process_cluster_dict_with_quality(clusters_dict, qualities)
        else:
            raise ValueError("Неизвестный метод группировки")
        
        # Вычисляем средние оценки для каждого кластера
        cluster_results = []
        for i, cluster in enumerate(clusters):
            answers_in_cluster = cluster['answers']
            embeddings_in_cluster = cluster.get('embeddings', [])
            
            # Вычисляем средние оценки
            avg_confidence = np.mean([a.confidence for a in answers_in_cluster])
            avg_chunk_score = np.mean([a.chunk_score for a in answers_in_cluster])
            combined_score = avg_confidence * avg_chunk_score
            
            # Вычисляем когерентность кластера
            if len(embeddings_in_cluster) > 1:
                cluster_distances = pairwise_distances(embeddings_in_cluster, metric='cosine')
                coherence = 1 - np.mean(cluster_distances)
            else:
                coherence = cluster.get('coherence', 1.0)
            
            cluster_results.append({
                'cluster_id': i,
                'answers': answers_in_cluster,
                'avg_confidence': avg_confidence,
                'avg_chunk_score': avg_chunk_score,
                'combined_score': combined_score,
                'coherence': coherence,
                'size': len(answers_in_cluster)
            })
        
        # Сортируем кластеры по комбинированной оценке
        cluster_results.sort(key=lambda x: x['combined_score'] * x['coherence'], reverse=True)
        
        return cluster_results
    
    def _process_cluster_dict(self, clusters_dict):
        """Преобразует словарь кластеров в унифицированный формат"""
        clusters = []
        for cluster_id, cluster_data in clusters_dict.items():
            answers_in_cluster, embeddings_in_cluster = zip(*cluster_data)
            clusters.append({
                'answers': list(answers_in_cluster),
                'embeddings': list(embeddings_in_cluster),
                'size': len(answers_in_cluster)
            })
        return clusters
    
    def _process_cluster_dict_with_quality(self, clusters_dict, qualities):
        """Преобразует словарь кластеров с учетом качества"""
        clusters = []
        for cluster_id, cluster_data in clusters_dict.items():
            answers_in_cluster, embeddings_in_cluster = zip(*cluster_data)
            clusters.append({
                'answers': list(answers_in_cluster),
                'embeddings': list(embeddings_in_cluster),
                'coherence': qualities.get(cluster_id, 0),
                'size': len(answers_in_cluster)
            })
        return clusters
    
    def select_final_answer(self, cluster_results, strategy='best_cluster'):
        """
        Выбор итогового ответа на основе группировки
        """
        if not cluster_results:
            return None
        
        if strategy == 'best_cluster':
            # Выбираем лучший ответ из лучшего кластера
            best_cluster = cluster_results[0]
            best_answer = max(best_cluster['answers'], 
                             key=lambda a: a.confidence * a.chunk_score)
            return best_answer
        
        elif strategy == 'weighted_combination':
            # Взвешенная комбинация ответов из топовых кластеры
            top_answers = []
            weights = []
            
            for cluster in cluster_results[:3]:  # Рассматриваем топ-3 кластера
                for answer in cluster['answers']:
                    weight = answer.confidence * answer.chunk_score * cluster['coherence']
                    top_answers.append(answer)
                    weights.append(weight)
            
            if not weights:
                return None
                
            # Нормализуем веса
            weights = np.array(weights) / sum(weights)
            
            # Выбираем ответ с максимальным весом
            best_idx = np.argmax(weights)
            return top_answers[best_idx]
        
        elif strategy == 'ensemble':
            # Ансамблевый подход
            answer_scores = defaultdict(float)
            
            for cluster in cluster_results:
                cluster_weight = cluster['combined_score'] * cluster['coherence']
                
                for answer in cluster['answers']:
                    answer_scores[answer.text] += cluster_weight * answer.confidence * answer.chunk_score
            
            if not answer_scores:
                return None
                
            # Выбираем ответ с наивысшей совокупной оценкой
            best_answer_text = max(answer_scores.items(), key=lambda x: x[1])[0]
            
            # Находим оригинальный объект ответа
            for cluster in cluster_results:
                for answer in cluster['answers']:
                    if answer.text == best_answer_text:
                        return answer
            
            return None

def process_answers_with_grouping(answers, grouping_method='adaptive', selection_strategy='best_cluster'):
    """
    Полный процесс обработки ответов с группировкой
    """
    # Инициализируем группер
    grouper = AnswerGrouper()
    
    # Группируем ответы
    clusters = grouper.group_answers(answers, method=grouping_method)
    
    # Выбираем итоговый ответ
    final_answer = grouper.select_final_answer(clusters, strategy=selection_strategy)
    
    # Дополнительная информация для анализа
    cluster_info = []
    for cluster in clusters:
        cluster_info.append({
            'cluster_id': cluster['cluster_id'],
            'size': cluster['size'],
            'avg_confidence': cluster['avg_confidence'],
            'avg_chunk_score': cluster['avg_chunk_score'],
            'combined_score': cluster['combined_score'],
            'coherence': cluster['coherence'],
            'answers': [answer.text for answer in cluster['answers']],
            'sample_answer': cluster['answers'][0].text[:100] + '...' if cluster['answers'] else ''
        })
    
    return final_answer, cluster_info
```
